```python
from download_with_aria2 import BiosinoOrgMetadataFile
import json
```

```python
def add_tool_to_json(input_file="../download_state.json", output_file="../download_state_aria2.json"):
    # Load the JSON data from file
    with open(input_file, "r") as file:
        data = json.load(file)

    # Add "tool": "lftp" to each entry
    for key in data:
        data[key]["tool"] = "lftp"
        data[key]["verified_with_md5"] = False
        data[key]["checksum_valid"] = False

    # Save the modified data to a new file
    with open(output_file, "w") as file:
        json.dump(data, file, indent=4)

# Run the function
add_tool_to_json()
```
